```python
from gensim.models import LdaModel
import gensim
import time
import pyLDAvis
pyLDAvis.enable_notebook()
import pyLDAvis.gensim
```

```python
corpus_tfidf = gensim.corpora.MmCorpus('corpus_tfidf.mm')
```

```python
gensim_dict =  gensim.corpora.dictionary.Dictionary.load('dictionary.dict')
```

```python
lda_tfidf = LdaModel.load('lda_10_topics.ldamodel')
```

```python
start = time.time()
lda_display = pyLDAvis.gensim.prepare(lda_tfidf, corpus_tfidf, gensim_dict)
end = time.time()
print((end-start)/3600,'hrs')
```

```python
pyLDAvis.display(lda_display)
```

```python
pyLDAvis.save_html(lda_display, 'topic_pyLDAvis.html')
```
